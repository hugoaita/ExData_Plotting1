```python
# Create dataframe from table with cols and index
def create_table(table, cols):

  t = pd.DataFrame(table[cols].unique(), columns=[cols])
  t.index.name = 'id_'+cols
  t.index += 1
  return t.reset_index()
```

```python
def assign_id_barrios(table, t, cols):
  # Replace the value from table with the id from the table t
  df = pd.merge(table, t, how='outer', on=[cols, cols])
  df.drop(cols, axis=1, inplace=True)
  return df
```
